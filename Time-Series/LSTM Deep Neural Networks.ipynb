## Data exploration

```python
df = pd.read_csv("transactions.csv")
df.info()
```

```python
df = df[df.store_nbr==1]
# Converting the column date into datetime format
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')
df = df.set_index('date')
df = df.drop(columns = ['store_nbr'])
```

```python
sns.lineplot(data = df, x=df.index, y='transactions', color = 'coral', linewidth=1.5, label = 'Transactions')
plt.grid('on', linewidth = 1, linestyle = ':', alpha= 0.5)
plt.xlabel('Time', fontsize = 15)
plt.ylabel('Transactions', fontsize = 15)
plt.tick_params(axis='x', labelsize=13)
plt.tick_params(axis='y', labelsize=13)
plt.subplots_adjust(left=0.16)
plt.subplots_adjust(bottom=0.24)
```
